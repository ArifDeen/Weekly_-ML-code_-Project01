```python
# Loading the data
import numpy as np
import pandas as pd
```

```python
cell = pd.read_csv("cell_phones.csv")
cell
```

```python
pop = pd.read_csv("population1.csv")
pop
```

```python
geo = pd.read_csv("ddf--entities--geo--country.csv")
geo
```

```python
# joining the data
merged = cell.merge(pop, how='inner')
merged
```

```python
rename_geo = geo.rename(columns={"country":"geo", "name":"Country"})
rename_geo
```

```python
joined = rename_geo.merge(merged, how='inner')
joined
```

```python
# Make features
joined['cell_phones_per_person'] = joined['cell_phones_total']/joined['population']
joined['cell_phones_per_person']
```

```python
usa = joined.loc[joined['geo'] == 'usa']
usa
```

```python
geo_upper = joined['geo'].apply(lambda x: x.upper())
geo_upper
```
